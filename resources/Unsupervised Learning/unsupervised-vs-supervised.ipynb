```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import matplotlib.pyplot as plt
```

```python
NUMBER_OF_DATA_POINTS = 100
```

```python
# Generating Random Data
from sklearn.datasets.samples_generator import make_blobs
X, Y = make_blobs(n_samples=NUMBER_OF_DATA_POINTS
                  , centers=4,
                       cluster_std=0.60, random_state=0)
```

```python
plt.scatter([x[0] for x in X],[x[1] for x in X])
plt.show()
plt.scatter([x[0] for x in X],[x[1] for x in X],c=Y)
plt.show()
```

```python
from sklearn.multioutput import MultiOutputRegressor
from sklearn.linear_model import LogisticRegression
clf = MultiOutputRegressor(LogisticRegression())
clf.fit(X,pd.get_dummies(pd.DataFrame(Y)).values )
```

```python
plt.scatter([x[0] for x in X],[x[1] for x in X], c=clf.predict(X))
```

```python
from sklearn.cluster import KMeans
clf = KMeans(n_clusters=4)
clf.fit(X)
```

```python
plt.scatter([x[0] for x in X],[x[1] for x in X],c=clf.predict(X))
```

```python
centers = clf.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
plt.scatter([x[0] for x in X],[x[1] for x in X])
```
