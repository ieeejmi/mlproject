```python
import numpy as np, random
import matplotlib.pyplot as plt

dX = np.random.rand(10000,1)
X = np.zeros((10000,2))
X[:,0] = dX.reshape(-1)
X[:,1] = (dX*2).reshape(-1)

y = np.mean(X, axis=1).reshape(-1,1) + np.random.rand(10000,1)
```

```python
X[:10]
```

```python
y[:10].reshape(-1,1)
```

```python
plt.scatter(np.arange(0,100), X[:100,0], marker='o')
plt.scatter(np.arange(0,100), X[:100,1], marker='x')
plt.title("Scatter plot of feature 1")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg = reg.fit(X, y)
```

```python
y_test = y
y_pred = reg.predict(X)
```

```python
plt.scatter(np.arange(0,100,1), y_test[:100], marker='x',
            color='red')
plt.plot(y_pred[:100])
plt.title("Model")
plt.show()
```

```python
from sklearn.metrics import r2_score, mean_squared_error
print('Mean Square Error: ', mean_squared_error(y_test, y_pred))
print('R2 Score: ', r2_score(y_test, y_pred))
```
